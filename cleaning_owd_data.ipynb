```python
df = pd.read_csv("open_energy_data.csv")
df.head()
```

```python
df_clean = df[df["year"]>2000]
```

```python
df_clean =df_clean.replace(np.nan, 0)
```

```python
df_clean.head()
```

```python
for colu in df_clean.columns:
    df_clean[colu]
    if colu != "country" and colu != "year" and colu != "iso_code":
        df_clean[colu] = df_clean[colu].astype(float)
df_clean.to_csv("open_energy_data_clean.csv",index=False)
```
